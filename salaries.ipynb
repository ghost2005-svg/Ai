```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import pandas as pd
df = pd.read_csv("Salaries.csv", low_memory=False)
df.head(10)
```

```python
df.describe()
```

```python
df.info()
```

```python
#THIS PLOT IS MADE FROM HELP OF AI
df['BasePay'] = pd.to_numeric(df['BasePay'], errors='coerce')
df['OvertimePay'] = pd.to_numeric(df['OvertimePay'], errors='coerce')
plt.scatter(df['BasePay'], df['OvertimePay'], alpha=0.2, color='purple')
plt.title('Relationship between Base Pay and Overtime')
plt.xlabel('Base Pay ($)')
plt.ylabel('Overtime Pay ($)')
plt.show()
```

```python
# THIS PLOT MADE BY MY SELF
avg_pay = df.groupby('Year')['BasePay'].mean()
avg_pay.plot(kind='bar')
plt.title('Average Salary per Year')
plt.xlabel('Year')
plt.ylabel('Base Pay')
plt.show()
```

```python
#ALSO THIS
df['BasePay'].plot(kind='box')
plt.title('Salary Distribution')
plt.ylabel('Base Pay')
plt.show()
```
