```
# Smart Student Record Analyzer (Beginner Simple Version)

filename = "students.txt"

def add_student():
    f = open(filename, "a")
    sid = input("Enter ID: ")
    name = input("Enter Name: ")
    age = input("Enter Age: ")
    grade = input("Enter Grade (A-F): ").upper()
    marks = input("Enter Marks (0-100): ")
    record = sid + "," + name + "," + age + "," + grade + "," + marks + "\n"
    f.write(record)
    f.close()
    print("Student added successfully!\n")

def view_students():
    try:
        f = open(filename, "r")
        data = f.readlines()
        f.close()
        if len(data) == 0:
            print("No records found.\n")
            return
        print("ID\tName\t\tAge\tGrade\tMarks")
        print("--------------------------------------------")
        for line in data:
            sid, name, age, grade, marks = line.strip().split(",")
            print(f"{sid}\t{name}\t\t{age}\t{grade}\t{marks}")
        print("--------------------------------------------")
        print("Total Students:", len(data), "\n")
    except FileNotFoundError:
        print("No file found. Add students first.\n")

def search_student():
    keyword = input("Enter ID or Name to search: ").lower()
    found = False
    try:
        f = open(filename, "r")
        for line in f:
            if keyword in line.lower():
                print("Record Found:", line.strip())
                found = True
        f.close()
        if not found:
            print("No record found.\n")
    except FileNotFoundError:
        print("No file found.\n")

def update_student():
    sid = input("Enter ID to update: ")
    lines = []
    found = False
    try:
        f = open(filename, "r")
        lines = f.readlines()
        f.close()
        f = open(filename, "w")
        for line in lines:
            data = line.strip().split(",")
            if data[0] == sid:
                print("Current Record:", line.strip())
                name = input("Enter New Name: ")
                age = input("Enter New Age: ")
                grade = input("Enter New Grade (A-F): ").upper()
                marks = input("Enter New Marks (0-100): ")
                new_line = sid + "," + name + "," + age + "," + grade + "," + marks + "\n"
                f.write(new_line)
                found = True
                print("Record Updated.\n")
            else:
                f.write(line)
        f.close()
        if not found:
            print("ID not found.\n")
    except FileNotFoundError:
        print("No file found.\n")

def delete_student():
    sid = input("Enter ID to delete: ")
    lines = []
    found = False
    try:
        f = open(filename, "r")
        lines = f.readlines()
        f.close()
        f = open(filename, "w")
        for line in lines:
            data = line.strip().split(",")
            if data[0] != sid:
                f.write(line)
            else:
                found = True
        f.close()
        if found:
            print("Record Deleted.\n")
        else:
            print("ID not found.\n")
    except FileNotFoundError:
        print("No file found.\n")

def analyze_data():
    try:
        f = open(filename, "r")
        data = f.readlines()
        f.close()
        if len(data) == 0:
            print("No records to analyze.\n")
            return
        marks_list = []
        for line in data:
            sid, name, age, grade, marks = line.strip().split(",")
            marks_list.append(int(marks))
        avg = sum(marks_list) / len(marks_list)
        highest = max(marks_list)
        lowest = min(marks_list)
        print("----- Data Analysis -----")
        print("Average Marks:", avg)
        print("Highest Marks:", highest)
        print("Lowest Marks:", lowest)
        print("-------------------------\n")
    except FileNotFoundError:
        print("No file found.\n")

def main():
    while True:
        print("====== Smart Student Record Analyzer ======")
        print("1. Add Student")
        print("2. View Students")
        print("3. Search Student")
        print("4. Update Student")
        print("5. Delete Student")
        print("6. Analyze Data")
        print("7. Exit")
        choice = input("Enter your choice: ")
        if choice == "1":
            add_student()
        elif choice == "2":
            view_students()
        elif choice == "3":
            search_student()
        elif choice == "4":
            update_student()
        elif choice == "5":
            delete_student()
        elif choice == "6":
            analyze_data()
        elif choice == "7":
            print("Thank you for using Smart Student Record Analyzer!")
            break
        else:
            print("Invalid choice. Try again.\n")

main()
```
